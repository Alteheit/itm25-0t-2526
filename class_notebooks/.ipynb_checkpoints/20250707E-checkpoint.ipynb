```python
# What is the quantity purchased per transaction?
# Stakeholder: Why are the transaction IDs like that? I want it in the form "tx-{id}"

df = pd.read_csv('denormalized_li.csv')

# What do you have?
df

# Step 1: Map transaction_id from the float it is to the string your boss wants
def clean_transaction_id(x):
    x = int(x)
    return f'tx-{x}'
df = df[df['transaction_id'].notnull()]

df['transaction_id'].apply(clean_transaction_id)

## The other way
df = pd.read_csv('denormalized_li.csv')

df = df[df['transaction_id'].notnull()]

df['transaction_id'] = df['transaction_id'].astype(int)
df['transaction_id'] = 'tx-' + df['transaction_id'].astype(str)

df.groupby(['transaction_id'])['line_item_quantity'].sum()
```
